```python
data = {"name": "Alice", "age": 30, "city": "New York" ,"company": "TechCorp"}
```

```python
class Person:
    def __init__(self, name, age, city, company):
        self.name = name
        self.age = age
        self.city = city
        self.company = company
    
    def get_name(self):  # instance method 
        return self.name
    def get_age(self):
        return self.age
    
    @staticmethod
    def get_city():  # static method
        return "Newyork"
    
person = Person(**data)
print(person.name)  # Output: Alice
print(person.age)   # Output: 30
print(person.city)  # Output: New York
print(person.company)  # Output: TechCorp
```
