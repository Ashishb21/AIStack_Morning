```python
from langchain_ollama import ChatOllama

llm =ChatOllama(model="llama3.2",
                temperature=0.9,  
                base_url="http://localhost:11434" )
```

```python
from langchain.agents import create_agent
agents=create_agent(llm, system_prompt="You are a helpful assistant that can answer questions about the world. You have access to the following tools:",tools=[])
```

```python
user_question="what is the current time ?"
agent_response=agents.invoke({"messages":[{"role":"user","content":user_question}]})
print(agent_response)
```

```python
agent_response['messages'][-1].content
```

```python
from datetime import datetime
from langchain.agents import create_agent
from langchain_core.tools import tool

@tool
def get_current_time():
    "This function returns the current time in a human-readable format."
    from datetime import datetime
    return datetime.now().strftime("%Y-%m-%d %H:%M:%S")

@tool
def get_weather():
    "This function returns the current weather in a specific location."
    # In a real implementation, you would call a weather API here.
    return "The current weather is sunny with a temperature of 25Â°C."

@tool
def get_db_record(record_id: int):
    "This function will return the record from db based on the id provided"
    print(f"Fetching record with ID: {record_id} from the database...")
    return f"Record {record_id}: This is a sample record from the database."

agents=create_agent(llm, system_prompt="You are a helpful assistant that can answer questions about the world. You have access to the following tools:",
                    tools=[get_current_time, get_weather, get_db_record])

user_question="give the db records for the id=1002 ?"
agent_response=agents.invoke({"messages":[{"role":"user","content":user_question}]})
print(agent_response["messages"][-1].content)
```
